### Creating Inflation Adjusted Budget and Gross

```python
#Adjust for inflation
import cpi
def inflation(df,col):
    return df.apply(lambda x: cpi.inflate(x[col],int(x.title_year)), axis=1)

movies['inf_adj_budget']=inflation(movies, 'budget')
movies['inf_adj_gross']=inflation(movies, 'gross')

movies.pop('budget')
movies.pop('gross')
```

```python
movies
```

### One Hot Encoding Genre

```python
genres=['Action','Adult','Adventure','Animation','Biography','Comedy','Crime','Documentary',
        'Drama','Family','Fantasy','Film Noir','Game Show','History','Horror','Musical','Music',
        'Mystery,News','Reality-TV','Romance','Sci-Fi','Short','Sport','Talk-Show','Thriller','War','Western'] #source:imdb
```

```python
#one hot code genres
for genre in genres:
    movies[genre]=0
for obs in zip(movies.index,movies['genres']):
    for genre in genres:
        if genre in obs[1]:
            movies[genre][obs[0]]=1
```

```python
movies.head()
```

```python
movies.pop('genres')
movies.hist(figsize=(15,15))
plt.subplots_adjust(hspace=0.8)
```

```python
plt.scatter(movies['inf_adj_budget'],movies['inf_adj_gross'])
plt.xlabel('(in $10 millions)')
plt.ylabel('(in $100 millions)')
print("correlation between the movie's budget and its gross is:",np.corrcoef(movies['inf_adj_budget'],movies['inf_adj_gross'])[0][1])
```

Surprisingly there isn't that much correlation between the budget and the gross!

### Combining GDP Growth Rate Data

Next, we are going to use the GDP growth rate in the US from 1929 to also use as one of our predictor variables, as GDP growth rate is a strong indicator of the nation's economic health.

```python
GDPchange=pd.read_csv('GDPchange.csv',delimiter='\t')
```

```python
GDPchange
```

```python
GDPchange.pop('Nominal GDP (trillions)')
GDPchange.pop('Real GDP (trillions)')
GDPchange.pop('Events Affecting GDP')
```

```python
movies=movies.merge(GDPchange,how='inner',left_on='title_year',right_on='Year')
movies.head()
```

```python
movies['GDP Growth Rate'].values
```

```python
movies[movies['GDP Growth Rate'].isnull()]
```

```python
#fill missing value at 1929 with 1930 data
movies['GDP Growth Rate'].fillna(value='-8.5%',inplace=True)
movies.iloc[[2995]]
```

```python
#turn percentage to float
for obs in zip(movies.index,movies['GDP Growth Rate']):
    movies['GDP Growth Rate'][obs[0]]=obs[1].replace('%','')
    movies['GDP Growth Rate'][obs[0]]=float(movies['GDP Growth Rate'][obs[0]])
    
```
